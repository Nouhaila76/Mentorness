```python
 # Calculate the percentage of active job cards
Total_Number_of_JobCards_issued = data['Total No. of JobCards issued'].sum()
Total_Number_of_Active_Job_Cards = data['Total No. of Active Job Cards'].sum()

# Calculate the job card utilization rate
job_card_utilization_rate = (Total_Number_of_Active_Job_Cards / Total_Number_of_JobCards_issued) * 100

# Display the job card utilization rate
print("job card utilization rate:", job_card_utilization_rate, "%")
```

```python
# Calculate the total number of workers and total number of active workers
Total_number_of_Workers = data['Total No. of Workers'].sum()
Total_number_of_Active_Workers = data['Total No. of Active Workers'].sum()

# Calculate the percentage of active workers
taux_workers_actif = (Total_number_of_Active_Workers / Total_number_of_Workers) * 100

# Display the results
print("Total number of workers:", Total_number_of_Workers)
print("Total number of active workers:", Total_number_of_Active_Workers)
print("Percentage of active workers:", taux_workers_actif, "%")
```

```python
# Calculate total works takenup, ongoing, and completed
total_works_takenup = data['Total No. of Works Takenup (New+Spill Over)'].sum()
total_works_ongoing = data['Number of Ongoing Works'].sum()
total_works_completed = data['Number of Completed Works'].sum()

# Calculate the ratio of completed works
ratio_completed = (total_works_completed / total_works_takenup) * 100

# Print the totals and the ratio
print("Total works undertaken:", total_works_takenup)
print("Total works ongoing:", total_works_ongoing)
print("Total works completed:", total_works_completed)
print("Ratio of completed works: {:.2f}%".format(ratio_completed))
```

```python
# Calculer le nombre moyen de jours de travail par mÃ©nage pour chaque Ã‰tat
state_avg_days = data.groupby('state_name')['Average days of employment provided per Household'].mean().reset_index()
state_avg_days.columns = ['State', 'Avg_Days_Worked']
print(state_avg_days)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the average number of days worked per household for each state
plt.figure(figsize=(12, 8))
sns.barplot(data=state_avg_days, x='Avg_Days_Worked', y='State', palette='viridis')
plt.title('Average Number of Days Worked per Household by State')
plt.xlabel('Average Number of Days Worked')
plt.ylabel('State')
plt.show()

```

```python


# Calculate the total expenditure
total_expenditure = data['Total Exp(Rs. in Lakhs.)'].sum()

# Calculate the expenditures in percentages
wages_percentage = (data['Wages(Rs. In Lakhs)'].sum() / total_expenditure) * 100
material_and_skilled_wages_percentage = (data['Material and skilled Wages(Rs. In Lakhs)'].sum() / total_expenditure) * 100
adm_expenditure_percentage = (data['Total Adm Expenditure (Rs. in Lakhs.)'].sum() / total_expenditure) * 100

# Use the sum of the percentage columns if they represent cumulative values
nrm_expenditure_percentage = data['percentage of NRM Expenditure(Public + Individual)'].mean()
category_b_works_percentage = data['percentage of Category B Works'].mean()
agriculture_expenditure_percentage = data['percentage of Expenditure on Agriculture & Agriculture Allied Works'].mean()

# Create a DataFrame for the percentages
expenditure_data = {
    'Category': ['Wages', 'Material and Skilled Wages', 'Administrative Expenditure', 
                 'NRM Expenditure', 'Category B Works', 'Agriculture & Allied Works'],
    'Percentage': [wages_percentage, material_and_skilled_wages_percentage, adm_expenditure_percentage, 
                   nrm_expenditure_percentage, category_b_works_percentage, agriculture_expenditure_percentage]
}

expenditure_df = pd.DataFrame(expenditure_data)

# Print the DataFrame to verify the calculations
print(expenditure_df)

# Create the pie chart
plt.figure(figsize=(10, 7))
plt.pie(expenditure_df['Percentage'], labels=expenditure_df['Category'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of NREGA Expenditures')
plt.show()
```

```python
# Distribution of average days of employment provided per household
plt.figure(figsize=(10, 6))
sns.histplot(data['Average days of employment provided per Household'], bins=30, kde=True)
plt.title('Distribution of Average Days of Employment Provided per Household')
plt.xlabel('Average Days of Employment')
plt.ylabel('Frequency')
plt.show()
```

```python
# Total households worked by state
households_worked = data.groupby('state_name')['Total Households Worked'].sum().sort_values(ascending=False)

plt.figure(figsize=(14, 8))
households_worked.plot(kind='bar')
plt.title('Total Households Worked by State')
plt.xlabel('State')
plt.ylabel('Total Households Worked')
plt.show()

# Total households completed 100 days of employment by state
households_100_days = data.groupby('state_name')['Total No of HHs completed 100 Days of Wage Employment'].sum().sort_values(ascending=False)

plt.figure(figsize=(14, 8))
households_100_days.plot(kind='bar')
plt.title('Total Households Completed 100 Days of Employment by State')
plt.xlabel('State')
plt.ylabel('Total Households Completed 100 Days')
plt.show()
```

```python
# SC and ST workers among active workers
sc_workers = data['SC workers against active workers'].sum()
st_workers = data['ST workers against active workers'].sum()
other_workers = data['Total No. of Active Workers'].sum() - (sc_workers + st_workers)

worker_demographics = pd.DataFrame({
    'Category': ['SC Workers', 'ST Workers', 'Other Workers'],
    'Count': [sc_workers, st_workers, other_workers]
})

plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Count', data=worker_demographics)
plt.title('Worker Demographics Among Active Workers')
plt.xlabel('Worker Category')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter plot of Approved Labour Budget vs Persondays of Central Liability
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Approved Labour Budget', y='Persondays of Central Liability so far', data=data)
plt.title('Approved Labour Budget vs Persondays of Central Liability')
plt.xlabel('Approved Labour Budget (in Lakhs)')
plt.ylabel('Persondays of Central Liability')
plt.show()

# Correlation coefficient
correlation = data[['Approved Labour Budget', 'Persondays of Central Liability so far']].corr().iloc[0, 1]
print(f'Correlation between Approved Labour Budget and Persondays of Central Liability: {correlation}')
```
